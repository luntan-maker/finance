```python
aa = yf.Ticker('aa')
aa = aa.history('1y')
aa.drop(columns=['Dividends', 'Stock Splits'], inplace=True)
aa['50MA'] = aa.Close.rolling(50).mean()
aa['200MA'] = aa.Close.rolling(200).mean()
aa.plot(y=["50MA", "200MA"], kind="line", figsize=(10, 10), use_index=True)
```

```python
# Ticker info and EMA
smallRange = 13
bigRange = 200
aa = yf.Ticker('GME')
aa = aa.history('1y')
aa.drop(columns=['Dividends', 'Stock Splits'], inplace=True)
aa['13EMA'] = aa['Close'].ewm(span=13, adjust=False).mean()
aa['48EMA'] = aa['Close'].ewm(span=48, adjust=False).mean()
aa['200EMA'] = aa['Close'].ewm(span=200, adjust=False).mean()
# aa.plot(y=["13EMA", "48EMA", "200EMA"], kind="line", figsize=(10, 10), use_index=True)
```

```python
price =  np.random.randint(250, size= 254)
```

```python
#RSI
delta = aa['Close'].diff()
up = delta.clip(lower=0)
down = -1*delta.clip(upper=0)
ema_up = up.ewm(com=13, adjust=False).mean()
ema_down = down.ewm(com=13, adjust=False).mean()
rs = ema_up/ema_down
aa["RSI"] = 100 - (100/ (1 + rs))
```

```python
# Call logic
aa['position1'] = ( aa['13EMA'] > aa['48EMA'] ) \
    & (price> aa["200EMA"]) & (aa["RSI"] > 50)#> additional_line1
aa['pre_position1'] = aa['position1'].shift(1)
aa.dropna(inplace=True) # dropping the NaN values
aa['call'] = np.where(aa['position1'] == aa['pre_position1'], False, True)
```

```python
price =  np.random.randint(250, size= 253)
```

```python
# Put logic
aa['position'] = ( aa['13EMA'] < aa['48EMA'] ) \
    & (price < aa["200EMA"]) & (aa["RSI"] < 50)#> additional_line1
aa['pre_position'] = aa['position'].shift(1)
aa.dropna(inplace=True) # dropping the NaN values
aa['put'] = np.where(aa['position'] == aa['pre_position'], False, True)
```

```python
fig = plt.figure()
ax1 = fig.add_subplot()
ax1.plot(aa["13EMA"], color = "m")
ax1.plot(aa["48EMA"], color ="c")
# ax1.plot(aa.index, additional_line1)
ax1.plot(aa["RSI"], color="g")
ax1.plot(aa.loc[aa["call"]]["13EMA"], "*", color="b")
ax1.plot(aa.loc[aa["put"]]["13EMA"], "*", color="r")
```
